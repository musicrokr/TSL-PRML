```python
all_tags = []
for website in all_websites:
    data = get(website)
    
    try:
        #Access all the relevant urls and store them in a list
        soup = BeautifulSoup(data,"html.parser")
        alltags = soup.findAll("a")
        all_tags.append(alltags)
    except:
        print('error for',website)
```

```python
urls = []
names = []
for tag_set in all_tags:
    for a_tag in tag_set:
        href = a_tag.get("href")
        print(href)
        try:
            backslashcount=href.count("/")
            print(backslashcount)
            if backslashcount > 2 and href.startswith("https://www.tnp.sg"):
                urls.append(href)
        except:
            print('error')
```

```python
for item in urls:
    item = item.rstrip()
    data = get(item)
    article_entry = pd.DataFrame({'link': [],'title': [],'category':[],'sub_category':[],'author':[],'date':[],'article': []})
    try:
        soup = BeautifulSoup(data,"html.parser")

        article_title = soup.find('h1', {"class":"story-headline"}).text
        print(article_title)

        if soup.find('div', {"class":"byline-name"}) is not None:
            article_author = soup.find('div', {"class":"byline-name"}).text
                                       
        if soup.find('div', {"class":"field field-name-field-custom-byline"}) is not None:
            article_author = soup.find('div', {"class":"field field-name-field-custom-byline"}).text
                                       
        if soup.find('div', {"class":"byline-name"}) is None and soup.find('div', {"class":"field field-name-field-custom-byline"}) is None:
            article_author='-'
        print(article_author)

        #article category
        print('URL',item)
        article_category=re.findall(r'.*\/',item)
        category=article_category[0]
        major_category= category.split('/')[3]
        sub_category = category.split('/')[4]
        print('Article Category:',major_category)

        datetime_stamp = soup.find('div',{'class': 'report-details'}).find('div')
        print(datetime_stamp.text)
        
        for script in soup(["script", "style"]):
            script.decompose() 
    
    
        article = soup.find('div', {"class":"body-copy"})
        article_text =article.text

        article_entry=article_entry.append({'link': item,'title': article_title,'category':major_category,'sub_category':sub_category,'author':article_author,'date':datetime_stamp.text,'article': article_text}, ignore_index=True)
        all_articles.append(article_entry)
    except:
        print('No data for',item)
```

```python
df = pd.concat(all_articles)
df
```
