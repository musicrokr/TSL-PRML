```python
all_tags = []
for website in all_websites:
    data = get(website)

    try:
        #Access all the relevant urls and store them in a list
        soup = BeautifulSoup(data,"html.parser")
        alltags = soup.findAll("a")
        all_tags.append(alltags)
    except:
        print('error for',website)
```

```python
urls = []
names = []
for tag_set in all_tags:
    for a_tag in tag_set:
        href = a_tag.get("href")
        try:
            if "http://www.asiaone.com/" in href:
                urls.append(href)
        except:
            print('error')
```

```python
article = pd.DataFrame({'link': [],'title': [],'topics': [],'category':[],'sub_category':[],'author':[],'date':[],'article': []})
all_articles = []
```

```python
for item in urls:
    item = item.rstrip()
    data = get(item)
    article_entry = pd.DataFrame({'link': [],'title': [],'topics': [],'category':[],'sub_category':[],'author':[],'date':[],'article': []})
    try:
        soup = BeautifulSoup(data,"html.parser")

        title = soup.find('meta', attrs={'property': 'og:title'})
        article_title = title["content"]
        print(article_title)

        author = soup.find('div', attrs={'class': 'single-field-item'})
        article_author = author.text
        article_author = article_author.strip()
        print(article_author)

        topics = soup.find('meta', attrs={'name': 'keywords'})
        article_topics = topics["content"]
        print(article_topics)

        #article category
        print('URL',item)
        article_category=re.findall(r'.*\/',item)
        category=article_category[0]
        major_category= category.split('/')[3]
        sub_category = category.split('/')[4]
        print('Article Category:',major_category)
        print('Article Category:',sub_category)

        publish_date = soup.find('div', attrs={'class': 'field field-name-field-publication-date field--type-datetime field--label-hidden field-item'})
        article_date = publish_date.text
        article_date = article_date.rstrip()
        print(article_date)
        
        for script in soup(["script", "style"]):
            script.decompose() 
    
        article = soup.find('div', attrs={'property': 'content:encoded'})
        if article is None:
            article = soup.find('div', attrs={'class': 'field field-name-body field--type-text-with-summary field--label-hidden field-item'})
        article_text =article.get_text()

        article_entry=article_entry.append({'link': item,'title': article_title,'topics': article_topics,'category':major_category,'sub_category':sub_category,'author':article_author.strip(),'date':article_date.rstrip(),'article': article_text}, ignore_index=True)
        all_articles.append(article_entry)
    except:
        print('No data for',item)
```

```python
df = pd.concat(all_articles)
df
```
