## Channel News Asia

```python
cnaList = pd.read_csv("./articles/william/articles-final.csv", header=None)
cnaList.rename(columns={1:'link'}, inplace=True)
cnaList.head()
```

```python
test = pd.read_csv("./articles/william/articles-txt/articles05txt", header=None,delimiter="\t",engine="python",error_bad_lines=False)
test
```

```python
', '.join(test[0])
```

```python
def get_text(filename): 
    try:
        article_text = pd.read_csv("./articles/william/"+filename, header=None,delimiter="\t",engine="python",error_bad_lines=False)
        text =', '.join(article_text[0])
        return text
        
    except:
        print('Unable to parse',filename)
```

```python
cnaList["article"] = cnaList.apply(lambda x: get_text(x["link"]), axis=1)
```
